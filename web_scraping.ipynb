```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
def get_title(soup):
    try:
        title=soup.find("h1",attrs={"class":'-fs20 -pts -pbxs'}).text.strip()
    except AttributeError:
        title=""
    return title
        
```

```python
def get_price_after(soup):
    try:
        price=soup.find("span", attrs={"class":'-b -ubpt -tal -fs24 -prxs'}).string.strip()
    except AttributeError:
        price=""
    return price
    
```

```python
def get_price_before(soup):
    try:
        price=soup.find("span", attrs={"class":'-tal -gy5 -lthr -fs16 -pvxs -ubpt'}).string.strip()
    except AttributeError:
        try:
            price=soup.find("span", attrs={"class":'-b -ubpt -tal -fs24 -prxs'}).string.strip()
        except AttributeError:
            price=""   
    return price
```

```python
def get_rating(soup):
    try:
        rating=soup.find("div",attrs={"class":'stars _m _al'}).text.strip()
    except AttributeError:
        rating=""
    return rating
```

```python
def get_seller_evaluation(soup):
    try:
        evaluation=soup.find("bdo",attrs={"class":'-m -prxs'}).text.strip()
    except AttributeError:
        evaluation=""
    return evaluation
        
```

```python
def get_shipping_speed(soup):
    try:
        speed=soup.find("div",attrs={"class":'-pas -bt -fs12'}).find("span",attrs={"class":'-m'}).text.strip()
    except AttributeError:
        speed=""
    return speed
```

```python
def get_nb_of_reviews(soup):
    try:
        nb=soup.find("h2",attrs={"class":'-fs14 -m -upp -pvm'}).text.strip()
    except AttributeError:
        nb=""
    return nb
```

```python
def get_availability(soup):
    try:
        ava=soup.find("p",attrs={"class":'-df -i-ctr -fs12 -pbs -gy5'}).string.strip()
    except AttributeError:
        try:
            ava=soup.find("p",attrs={"class":'-df -i-ctr -fs12 -pbs -yl7'}).string.strip()
        except AttributeError:
            try:
                ava=soup.find("p",attrs={"class":'-df -i-ctr -fs12 -pbs -rd5'}).text.strip()
            except AttributeError:
                ava=""
    return ava
```

```python
import numpy as np
```

```python
if __name__=='__main__':
    
    HEADERS = ({'User-Agent':'https://useragentstring.com/Safari7.0.3_id_19858.php', 'Accept-Language': 'en-US, en; q=0.5'})
    URL="https://www.jumia.ma/catalog/?q=gaming"
    webpage = requests.get(URL, headers=HEADERS)
    soup=BeautifulSoup(webpage.content,'html.parser')
    links = soup.find_all("a",attrs={'class':'core'})
    
    links_list=[]
    for link in links:
        links_list.append(link.get("href"))
        
    d={'title':[],'price_before':[],'price_after':[],'rating':[],'seller_evaluation':[],'shipping_speed':[],
       'nb_reviews':[],'availability':[]}
    
    for link in links_list:
        new_web_page=requests.get("https://www.jumia.ma"+link,headers=HEADERS)
        new_soup=BeautifulSoup(new_web_page.content,'html.parser')
        
        d['title'].append(get_title(new_soup))
        d['price_before'].append(get_price_before(new_soup))
        d['price_after'].append(get_price_after(new_soup))
        d['rating'].append(get_rating(new_soup))
        d['seller_evaluation'].append(get_seller_evaluation(new_soup))
        d['shipping_speed'].append(get_shipping_speed(new_soup))
        d['nb_reviews'].append(get_nb_of_reviews(new_soup))
        d['availability'].append(get_availability(new_soup))
        
        
    jumia_df = pd.DataFrame.from_dict(d)
    jumia_df['title'].replace('', np.nan, inplace=True)
    jumia_df = jumia_df.dropna(subset=['title'])
    jumia_df.to_csv("jumia_data.csv", header=True, index=False)
        
        
   
    
```
