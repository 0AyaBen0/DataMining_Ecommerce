```python
def generer_utilisateur_factice(user_id, sexe=None):
    if sexe is None:
        sexe = random.choice(['Homme', 'Femme'])

    if sexe == 'Homme':
        prenom = random.choice(prenoms_hommes)
        categories_possibles = [cat for cat in categories if cat != 'Beauté']
    else:
        prenom = random.choice(prenoms_femmes)
        categories_possibles = categories

    categorie_preferee = random.choice(categories_possibles)
    last_product_purchased = random.choice(products[categorie_preferee]) if categorie_preferee in products else None

    nom = random.choice(noms_de_famille_marocains)
    nom_complet = f"{prenom} {nom}"
    email = f"{prenom.lower()}.{nom.lower()}{random.choice(domaines_email)}"

    return {
        "Identifiant": user_id,
        "Nom": nom_complet,
        "Email": email,
        "Âge": random.randint(18, 70),
        "Ville": random.choice(villes_marocaines),
        "Sexe": sexe,
        "Date d'inscription": fake.date_this_decade(),
        "Dernière connexion": fake.date_time_this_year(),
        "Catégorie préférée": categorie_preferee,
        "Appareil": random.choice(appareils),
        "produit acheté": last_product_purchased
    }
```

```python
utilisateurs = [generer_utilisateur_factice(user_id) for user_id in range(1, 501)]
```
